```python
PHIX = [4,6,8,14]
PSIX = [6,8,14,16]
```

Given the coordinates (numpy array), a reference ProDy AtomGroup, and the number of coordinates, return the $\Phi$ and $\Psi$ angles as a 2-tuple of 1D arrays of length $count$.

```python
def calc_dih(coords, reference, count):
    cnf  = reference.copy()
    phis = np.nan * np.zeros(count)
    psis = np.nan * np.zeros(count)
    for i, coord in enumerate(coords):
        cnf.setCoords(coord)
        phis[i] = prody.measure.calcDihedral(*[cnf[j] for j in PHIX])
        psis[i] = prody.measure.calcDihedral(*[cnf[j] for j in PSIX])

    return phis, psis
```

Given a glob pattern return the $\Phi$ and $\Psi$ angles of the pdb files matching the pattern.

```python
def load_pdb_dih(pattern):
    paths = glob.glob(pattern)
    ref   = prody.parsePDB(paths[0])
    crds  = itertools.imap(lambda p: prody.parsePDB(p).getCoords(), paths)
    return calc_dih(crds, ref, len(paths))
```

Given a glob pattern matching $XTC$ files, the path to a reference PDB file, and an optional subsampling count, return the dihedral angles.

```python
def load_xtc_dih(pattern, pdbpath, subsample=None):
    def load_coords():
        coords = list()
        for xtc in glob.iglob(pattern):
            trj = msmb.Trajectory.load_from_xtc(xtc, PDBFilename=pdbpath)
            if subsample:
                trj.subsample(subsample)
            for xyz in trj['XYZList']:
                coords.append(xyz)
        return coords
    ref  = prody.parsePDB(pdbpath)
    crds = load_coords()
    return calc_dih(crds, ref, len(crds))
```

Given the path to the $AWE$ cell definitions file and a ProDy AtomGroup as topology, return the dihedral angles.

```python
def load_cells_dih(cellspath, ref):
    with open(cellspath) as fd:
        ncells = int(fd.readline().strip().split()[-1])
        ncoord = int(fd.readline().strip().split()[-1])
        ndim   = int(fd.readline().strip().split()[-1])
        fd.readline()
        
        coords = np.zeros(ncells * ncoord * ndim)
        for i, line in enumerate(itertools.imap(str.strip, fd)):
            val = float(line)
            coords[i] = val
        coords = coords.reshape((ncells, ncoord, ndim))
        return calc_dih(coords, ref, ncells)
```

# Cell Coordinates

Calculate the $\Phi$ and $\Psi$ dihedral angles for the cluster centers.

```python
phis, psis = load_cells_dih('cells.dat', ala)
```

Compute the cell weights.

```python
def extractWeights(path, ncells):
    weights = np.zeros(ncells)
    with open(path) as fd:
        for i, line in enumerate(itertools.imap(str.strip, fd)):
            if line.startswith('#'): continue
            content = line.split(',')
            cellid = int(content[2])
            weight = float(content[3])
            weights[cellid] += weight
    return weights
```

Duplicate the $\Phi$ and $\Psi$ angles on either side (up and down) from the original data points to account for periodicity.

```python
assert len(phis) == len(psis), 'phis[%s] =/= psis[%s]' % (len(phis), len(psis))
P = np.array([phis, psis]).T
```

```python
P2 = np.zeros((9*len(P), 2))
for x in [-1, 0, 1]:
    for y in [-1, 0, 1]:
        offset = ((x+1)*3+y+1)*len(P)
        for c in xrange(len(P)):
            ph, ps = P[c,0] + x*360, P[c,1] + y*360
            ix = offset + c
            P2[ix, :] = [ph, ps]
```

```python
cells = vn.voronoi(P2.T[0], P2.T[1])
patches = list()
for c in cells:
    polygon = Polygon(c, True)
    patches.append(polygon)
```

Plot the cell centers as points with the voronoi boundaries colored by weight (free energy).

```python
plt.rc('font', size=16)

weights_msm = np.loadtxt('Data/Populations.dat')
weights_awe = extractWeights('walker-weights.csv', 100)

colors_msm = -np.log(weights_msm)
colors_awe = -np.log(weights_awe)
cmin = int(round(min(colors_awe.min(), colors_msm.min())))
cmax = int(round(max(colors_awe.max(), colors_msm.max()))) + 1

norm = plt.Normalize(cmin, cmax)


for name, weights, colors in [('msm', weights_msm, colors_msm), ('awe', weights_awe, colors_awe)]:
    fig = plt.figure()
    plt.scatter(P2.T[0], P2.T[1], marker='o', c=np.tile(colors, 9), linewidths=0.5, cmap=plt.cm.jet_r, norm=norm)
    plt.clim(cmin, cmax)
    patch_cltn = PatchCollection(patches, cmap=plt.cm.jet_r, alpha=.07, edgecolor=[1,1,1], linewidths=0, norm=norm)
    patch_cltn.set_array(colors)
    plt.gca().add_collection(patch_cltn, autolim=False)
    bar = plt.colorbar(ticks=range(cmin, cmax, 2))
    plt.clim(cmin, cmax)
    bar.set_label('Free Energy')
    plt.xlabel('$\Phi$')
    plt.ylabel('$\Psi$')
    plt.axis([-180,180,-180,180])
    plt.xticks(range(-180, 180+60, 60))
    plt.yticks(range(-180, 180+60, 60))
    plt.tight_layout()
    plt.show()

    # IPython Notebook cannot embed pdf images, so also save a png
    for suff in ['pdf', 'png']:
        outpath = 'ala-rama-%s.%s' % (name, suff)
        fig.savefig(outpath, bbox_inches='tight')
```
